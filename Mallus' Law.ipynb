```python
from sympy import Symbol 
from sympy import symbols 
from sympy import *
theta, phi, i, pi, E_x, E_y = symbols("theta, phi, i, pi, E_x, E_y",real=True)
```

```python
# input two matrices

#South path (Horizontally Polarized)
E = Matrix([[E_x],
     [E_y]])

A = Matrix([[0, 0],
     [0,1]])

B = Matrix(
    [[cos(2*theta),-sin(2*theta)],
     [-sin(2*theta),-cos(2*theta)]])
D = Matrix(
    [[0,0],
     [0,1]])
 
# This will return dot product
M1 = D*B*A*E

M1
```

```python
M2 = Transpose(M1)
M2
```

```python
Final = M2*M1
Final
E_final = Final[0]
```

```python
from sympy import simplify, cos, sin, expand, re, im
from sympy.plotting import plot

#Simplify expression 
result = simplify(E_final)
result = expand(result)

#Take the real part of the imaginary result
r = re(result)
r
```

```python

#return numerical values for symbols to process data
r1 = r.subs(E_y,1)

#magnitude of vector requires square root of dot product between matrices
#Intensity is 1/2*epsilon_0*c*E^2

p1 = plot(r1,(theta,0,sp.pi/2))
p1
```

```python


  
# Generate line graph
theta = np.linspace(0,120,100)
y1 = 1560*(np.cos(2*(theta*np.pi/180-5*np.pi/35))**2)

y_error = 20
x_error = 1
  

plt.rcParams['text.usetex'] = True
plt.figure(figsize=(15,10))
plt.plot(theta, y1)
plt.scatter(x,y,marker="*",s=1,edgecolors="black",c="yellow")
plt.errorbar(x, y,
             yerr = y_error,
             xerr = x_error,
            fmt='.')
plt.title('Power vs. Angle - Mallus Law')
plt.xlabel(r'Angle ($\theta$)')
plt.ylabel('Power (mV)')
plt.show()
```
