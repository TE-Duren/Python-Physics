```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set parameter values
alpha_1 = 0.025 # Pro-inflammatory background production
alpha_2 = 5 # ||PIC|| production (bars represent magnitude)
alpha_3 = 0.5 #PIC Concentration at which AIC concentration is half maximal
alpha_4 = 3.5 # ||AIC|| production
gamma = 1.25 # Clearance rate of PIC vs. AICs

# Build curves for plot
p = np.linspace(0, 1.25, 1000)
y = np.sqrt((((p**2)*(alpha_1 + alpha_2) + alpha_1)/(gamma * p * (1 + p**2))) - 1)
z = (alpha_4 * p**2)/(alpha_3**2 + p**2)

# Generate plot
plt.figure(figsize=(20,10))
plt.plot(p, y, color = 'crimson', linewidth = 3)
plt.plot(p, z, color = 'lightseagreen')
plt.grid(True)
plt.xlabel('Pro-inflammatory cytokine concentration', fontsize=12)
plt.ylabel('Anti-inflammatory cytokine concentration', fontsize=12)
plt.legend(('$dp/dt = 0$', '$da/dt = 0$'), loc="upper left", fontsize=18);
```
