```python
import numpy as np
import math
import matplotlib.pyplot as plt
import sympy as sp
from sympy import *
sp.init_printing()

#parameters
p,k,e,r,theta,x,t,u,r0 = sp.symbols('p k e r \theta x t u \r_0')


k=.5

r0=1
p=3000/(np.pi*r0**2)
e=np.e

U_r = p*e**(-k*r)
U_cf = p*(1-k*r)



y = integrate(r*U_cf,r)



xs = np.linspace(.14, .15, 1)  # [-2.   -0.75  0.5   1.75  3.  ]
yvals = [U_cf.subs(r, xi) for xi in xs]


plot(U_r,U_cf,y,(r,0,1),xlabel='r',legend=False,markers=[{'args': [xs, yvals, 'ro']}])
print(y)
```

```python
from sympy.plotting import plot3d
plot3d(-y, (theta, 0, 2*np.pi), (r, 0, .2))
y
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(projection='3d')

# Create the mesh in polar coordinates and compute corresponding Z.
r = np.linspace(0, 2.3, 1000)
p = np.linspace(0, 2*np.pi, 1000)
R, P = np.meshgrid(r, p)
Z = -159.16*(R**3)+477.73*(R**2)

# Express the mesh in the cartesian system.
X, Y = 1+R*np.cos(P), 2+R*np.sin(P)

# Plot the surface.
ax.plot_surface(X, Y, Z, cmap=plt.cm.YlGnBu_r)


# Tweak the limits and add latex math labels.
ax.set_zlim(0, 1000)
ax.set_ylim(2.2, -2.2)
ax.set_xlim(2.2, -2.2)
ax.set_xlabel(r'$\phi_\mathrm{real}$')
ax.set_ylabel(r'$\phi_\mathrm{im}$')
ax.set_zlabel(r'$F(r)$')
for angle in range(60,120):
    ax.view_init(20, angle)
    plt.draw()
    plt.pause(.001)

plt.show()
```
