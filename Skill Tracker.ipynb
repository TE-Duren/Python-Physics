```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#Mechanical Engineering

def x1(a,b,c,d):
    #The parameters below may are the major factors that improve your specialty
    #i.e. learning business management, music, writing, sports, etc.
    
    #Parameters - weights calculated through trial-and-error approximation
    #a = number of real-world tasks completed with specific skillset
    #b = number of new theoretical problems completed
    #c = number of practice sets on 1 targeted subject (i.e. Valuing stocks)
    #d = number of different subjects areas read
    
    x = 0.01*a + 0.005*b + 0.0025*c + 0.00125*d
    return x

#Ex: Below we are passing 500 real-world tasks completed within your
#specialty into the function. 
#which will have the most weight on the test output below (13.75)
#Ensure all inputs start at 0 for a new specialty being developed

x1 = x1(84,54,55,17)
skill_lvl1 = round(x1*10)
print(skill_lvl1)
```

```python
# Build curves for plot
# Linspace scales from 0 to 10, with 1000 points in between

x = np.linspace(0, 100, 10000)

#Sigmoid function is conventionally used to simulate artificial neuron activation,
#which maps the learning curve for any skillset of your choosing may be tracked to
#see the progress you are making. Offset by 5 to normalize the function

y = 1/(1+np.e**(-.1*(x-50))) 

# Generate plot
plt.figure(figsize=(20,10))

#Markers_on function below plots marker on where your mastery currently resides 

markers_on = [round(x1*1000)]

plt.plot(x, y, '-gD', markevery=markers_on)
plt.grid(True)
plt.xlabel('Skill Level', fontsize=12)
plt.ylabel('Percent Mastery', fontsize=12)
plt.title('Mechanical Engineering', fontsize=20)
```
